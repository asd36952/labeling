```python
import pickle
import numpy as np
import html

data = "../data/ANGELIS_POSITION.pkl"

with open(data, "rb") as f:
    data = pickle.load(f)
```

```python
relation_list = []

for sent, key, key_s, key_e, ans, ans_s, ans_e, rel, conf in data:
    relation_list.append(rel)
    
relation_list = sorted(list(set(relation_list)))
```

```python
relation_list
```

```python
relation_cnt_dict = {}
for rel in relation_list:
    relation_cnt_dict[rel] = 0
    
for sent, key, key_s, key_e, ans, ans_s, ans_e, rel, conf in data:
    relation_cnt_dict[rel] += 1
```

```python
relation_dict = {}
for rel in relation_list:
    relation_dict[rel] = 0
```

```python
data_dict = {}

for rel in relation_list:
    data_dict[rel] = []
    
idx_list = list(range(len(data)))
np.random.shuffle(idx_list)

for idx in idx_list:
    sent, key, key_s, key_e, ans, ans_s, ans_e, rel, conf = data[idx]
    data_dict[rel].append([html.unescape(sent), html.unescape(key), html.unescape(ans)])
```

```python
correct = 0
wrong = 0

wrong_idx_list = []

for i in range(100):
    target = np.random.choice(relation_list)

    if ((i + 1) % 3 == 0):
        tmp = target
    else:
        while(relation_dict[target] >= relation_cnt_dict[target]):
            target = np.random.choice(relation_list)

        tmp = np.random.choice(relation_list)
        while (tmp == "no_relation")|(target[:3] != tmp[:3]):
            tmp = np.random.choice(relation_list)
    
    tmp_data = data_dict[target][relation_dict[target]]
    print("Query %d" % (i + 1))
    print("Sentence:\n%s\nKeyword:\n%s\nTarget:\n%s\nRelation:\n%s\n" % (tmp_data[0], tmp_data[1], tmp_data[2], tmp))
    output = input("Is it True? 1 for True, 0 for False")
    
    if (int(output) == 1) & (target == tmp):
        correct += 1
    elif (int(output) == 0) & (target != tmp):
        correct += 1
    else:
        wrong += 1
        wrong_idx_list.append([tmp_data[0], tmp_data[1], tmp_data[2], target, tmp])
    relation_dict[target] += 1
        
print("Score = %d/%d = %f" % (correct, correct + wrong, correct/(correct + wrong)))
    
```
